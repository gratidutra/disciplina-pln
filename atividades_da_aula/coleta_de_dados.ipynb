```python
import chromedriver_autoinstaller
import time

import pandas as pd
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
```

```python
chromedriver_autoinstaller.install()

chrome_options = Options() 
chrome_options.add_argument("--headless")
driver = webdriver.Chrome(options=chrome_options)

url = "https://wp.ufpel.edu.br/computacao/ppgc/docentes/"

driver.get(url)

elem = driver.find_element_by_xpath('//*[@id="content"]/div[2]')

professor_list = []
all_li = elem.find_elements_by_tag_name("li")
for li in all_li:
    text = li.text
    professor_list.append(text)
    #print (text)

driver.quit()

professor_list
```

```python
projects = {} 

chromedriver_autoinstaller.install()
chrome_options = Options()
chrome_options.add_argument("--headless")

for professor in professor_list:
    print(professor)

    driver = webdriver.Chrome(options=chrome_options)
    
    try:
        url = "https://institucional.ufpel.edu.br/servidores"
        driver.get(url)
        time.sleep(3)
        
        search = driver.find_element(
            By.XPATH,
            '//*[@id="DataTables_Table_0_filter"]/label/input'
        )
        search.send_keys(professor)
        search.send_keys(Keys.RETURN)
        
        WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable(
                (By.XPATH, '//*[@id="DataTables_Table_0"]/tbody/tr/td')
            )
        ).click()
        
        WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable((By.XPATH, '//*[@id="proj-sup"]'))
        ).click()
        
        projects_table = driver.find_element(By.TAG_NAME, "table")
        links = projects_table.find_elements(By.TAG_NAME, "a")
        
        hrefs = []
        for link in links:
            href = link.get_attribute("href")
            if href and "projetos" in href:
                hrefs.append(href)
        
        projects[professor] = []
        
        for project_url in hrefs:
            driver.get(project_url)
            time.sleep(2)
            
            project_data = {
                'link': project_url,
                'resumo': None
            }
            
            try:
                resumo_element = driver.find_element(
                    By.XPATH, 
                    '//*[@id="conteudo"]/div[1]/div[1]/div[14]'
                )
                project_data['resumo'] = resumo_element.text.strip()
            except:
                print(f"Não foi possível obter o resumo do projeto: {project_url}")
            
            projects[professor].append(project_data)
    except:
        print("^^^^Professor Don't found")
    
    finally:
        driver.quit()
```
