### LowerCase

```python
def tolower(column_data):
    if not isinstance(column_data, str):
        return column_data  # Retorna como está se não for uma string

    # Converte cada item da coluna para minúsculas
    lower_text= column_data.lower()

    return lower_text

def remove_accents(text):
  if isinstance(text, str):
    return unidecode(text)
  else:
    return text

def remove_special_char(text):
    if isinstance(text, str):
        return re.sub(r'[^a-zA-Z0-9]', ' ', text)
    return text
```

```python
data = (
    data
    .pipe(lambda df: df.assign(tolower=df["review_text"].apply(tolower)))
    .pipe(lambda df: df.assign(remove_accents=df["tolower"].apply(remove_accents)))
    .pipe(lambda df: df.assign(texto_normalizado=df["remove_accents"].apply(remove_special_char)))
)

data[['review_text', 'texto_normalizado']]
```

# Exercicío 2

```python
## tokenizer
def tokenizer(text):

  if not isinstance(text, str):
    return text

  sentences = nltk.sent_tokenize(text)

  all_sentences = []
  for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    all_sentences.append(words)

  return all_sentences

## StopWords with spacy
def remove_stopwords_spacy(all_sentences):
  pt = spacy.load('pt_core_news_sm')
  stopwords_spacy_pt = pt.Defaults.stop_words

  for i in range(len(all_sentences)):
    all_sentences[i] = [word for word in all_sentences[i] if word not in stopwords_spacy_pt]

  return all_sentences

## short words
def clean_sentences(sentences):
      return [
        [word for word in sentence if len(word) >= 3]
       for sentence in sentences
    ]
```

```python
data = (
    data.head(100)
    .pipe(lambda df: df.assign(tokenized_text=df["review_text"].apply(tokenizer)))
    .pipe(lambda df: df.assign(text_without_stopwords=df["tokenized_text"].apply(remove_stopwords_spacy)))
    .pipe(lambda df: df.assign(tokens=df["text_without_stopwords"].apply(clean_sentences)))
)
```

## Exercicio 3

```python
text = '''
Joãozinho e Maria, dois irmãos aventureiros, decidiram explorar a
floresta misteriosa próxima à sua casa. A floresta era densa, com árvores altas e
folhagem espessa. Enquanto caminhavam, João e Maria encontraram pegadas
estranhas no chão e decidiram segui-las. As pegadas os levaram a uma clareira
onde descobriram uma casa feita inteiramente de doces! Ficaram fascinados e não
resistiram à tentação de experimentar os doces. Mas, assim que começaram a
comer, uma bruxa má e enrugada apareceu. A bruxa riu de forma maligna e os
assustou. Joãozinho e Maria, apavorados, deixaram os doces e fugiram da casa. A
bruxa tentou segui-los, mas eles eram ágeis e conseguiram escapar. Finalmente,
após uma longa e cansativa jornada, João e Maria voltaram em segurança para
sua casa, onde prometeram nunca mais se aventurar naquela floresta perigosa.
'''

def highlight_joao_maria(text):
    text = re.sub(r'\bJoão\b', 'JOÃO', text, flags=re.IGNORECASE)
    text = re.sub(r'\bMaria\b', 'MARIA', text, flags=re.IGNORECASE)
    return text

```
